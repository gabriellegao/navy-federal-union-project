```python
from data import load_normalized_dataset, extract_features
import pandas, numpy
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
df = load_normalized_dataset()
```

```python
df.head(10)
```

```python
# Most common words

doc_term_mat, co_occ_mat, count_vec = extract_features(df.NormalizedMessage.astype(str))
#numpy.shape(doc_term_mat.sum(axis=0))
top_term_freqs = pandas.DataFrame(
    columns=['token', 'freq'],
    data={
        'token': count_vec.get_feature_names(),
        'freq': numpy.array(doc_term_mat.sum(axis=0))[0]
        }
)
top_term_freqs.sort_values(by='freq', ascending=False).head(10)
```

```python
# Most common words filtering out stopwords

doc_term_mat, co_occ_mat, count_vec = extract_features(df.NormalizedMessage.astype(str), stop_words=stop_words)
#numpy.shape(doc_term_mat.sum(axis=0))
top_term_freqs = pandas.DataFrame(
    columns=['token', 'freq'],
    data={
        'token': count_vec.get_feature_names(),
        'freq': numpy.array(doc_term_mat.sum(axis=0))[0]
        }
)
top_term_freqs.sort_values(by='freq', ascending=False).head(100)
```

```python
#Most frequent bigrams filtering out stopwords

doc_term_mat, co_occ_mat, count_vec = extract_features(df.NormalizedMessage.astype(str), stop_words=stop_words, ngram_range=(2,2))
#numpy.shape(doc_term_mat.sum(axis=0))
top_term_freqs = pandas.DataFrame(
    columns=['token', 'freq'],
    data={
        'token': count_vec.get_feature_names(),
        'freq': numpy.array(doc_term_mat.sum(axis=0))[0]
        }
)
top_term_freqs.sort_values(by='freq', ascending=False).head(100)
```

```python
#Most frequent trigrams filtering out stopwords

doc_term_mat, co_occ_mat, count_vec = extract_features(df.NormalizedMessage.astype(str), stop_words=stop_words, ngram_range=(3,3))
#numpy.shape(doc_term_mat.sum(axis=0))
top_term_freqs = pandas.DataFrame(
    columns=['token', 'freq'],
    data={
        'token': count_vec.get_feature_names(),
        'freq': numpy.array(doc_term_mat.sum(axis=0))[0]
        }
)
top_term_freqs.sort_values(by='freq', ascending=False).head(100)
```

```python
# Most common cooccurs filtering out stopwords

doc_term_mat, co_occ_mat, count_vec = extract_features(df.NormalizedMessage.astype(str), stop_words=stop_words)

tokens = count_vec.get_feature_names()
cofreq_terms = [(x,y) for x in tokens for y in tokens]
cofreq = []
top_term_cofreqs = pandas.DataFrame(columns=['tokens', 'cofreq'], data={
    'tokens': cofreq_terms,
    'cofreq': [co_occ_mat[count_vec.vocabulary_[x],count_vec.vocabulary_[y]] for (x,y) in cofreq_terms]
})
```

```python
def load_subjectivity_lexicon():
    posUrl, negUrl = "../resources/sentiment_lexicon/positive-words.txt", "../resources/sentiment_lexicon/negative-words.txt"
    with open(posUrl, "r") as posFile, open(negUrl, "r") as negFile:
        posText, negText = posFile.read(), negFile.read()
        posLines, negLines = posText.split("\n"), negText.split("\n")
        return posLines[31:], negLines[31:]
```
